```python
#Creating subset of the dataframe with the first one million rows.
df = df_ords_prods_merge[:1000000]
```

```python
# Definition new function called price_label

def price_label(row):
  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
df_ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
# 0 stands for saturday. Therefore saturday is the busiest day of the week. 
```

## Step 2 - new column for busiest days

```python
# Create new if-statement for-loop for new requirements
result_v2 = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0 or value == 1:
    result_v2.append("Busiest days")
  elif value == 4 or value == 3:
    result_v2.append("Least busy days")
  else:
    result_v2.append("Regularly busy")
```

```python
result_v2
```

```python
# Add result to dataframe in new column
df_ords_prods_merge['Busiest days'] = result_v2
```

```python
# Check
df_ords_prods_merge.head()
```

## Step 3 Check & Obervations

```python
df_ords_prods_merge['Busiest days'].value_counts(dropna = False)
```

###### The sum for the busiest days equals the sum of the values 0 and 1, and the values for the least busy days add up to the sum of the values 3 and 4, confirming that the loop is working correctly. This is because the original frequency counts of the individual "groupings" add up.

## Step 4 Create a new column containing these labels called “busiest_period_of_day.”

```python
# Check frequency of values in the "order_hour_of_day" column 
df_ords_prods_merge['order_hour_of_day'].value_counts()
```

#### The client wants these grouped into three different segments. Therefore we can group the 24 hours of the day as follows: Most orders = 10, 11, 14, 15, 13, 12, 16, 9; Average orders = 17, 8, 18, 19, 20, 7, 21, 22; Fewest orders = 23, 6, 0, 1, 5, 2, 4, 3


```python
# Create a for-loop if statement using the above breakdown
hour_of_day = []

for value in df_ords_prods_merge["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    hour_of_day.append("Most orders")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    hour_of_day.append("Fewest orders")
  else:
    hour_of_day.append("Average orders")
```

```python
# Add result to dataframe as "busiest_period_of_day" column
df_ords_prods_merge['busiest_period_of_day'] = hour_of_day
```

```python
# Check
df_ords_prods_merge[['order_hour_of_day', 'busiest_period_of_day']].head(10)
```

## Step 5 - Print Frequency

```python
# check the frequency of the new column busiest preiod of day
df_ords_prods_merge['busiest_period_of_day'].value_counts()
```

###### These frequencies match the sums of the above "ungrouped" frequencies, confirming the correct application of the labels
