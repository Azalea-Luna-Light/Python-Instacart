### Sampling Data

```python
#creating samples / a list holding True/False values to the test np.random.rant() <= 0.7
np.random.seed(4)
dev = np.random.rand(len(df_ords_prods_merge)) <= 0.7
```

```python
dev
```

```python
np.random.rand(10)
```

```python
# Store 70%of the sample in the dataframe big
big = df_ords_prods_merge[dev]
```

```python
# Store 70%of the sample in the dataframe small
small = df_ords_prods_merge[~dev]
```

```python
len(df_ords_prods_merge)
```

```python
len(big) + len(small)
```
